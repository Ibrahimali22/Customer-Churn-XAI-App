# Telecom Customer Churn - EDA
This notebook documents the Exploratory Data Analysis (EDA) and Data Visualizations used to derive Business Insights.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
df.head()
```

```python
# 2. Churn Distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Churn', palette='pastel')
plt.title('Overall Customer Churn Rate')
plt.show()
```

```python
# 3. Impact of Contract Type on Churn
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Contract', hue='Churn', palette='viridis')
plt.title('Churn Rate strictly dictated by Contract Duration')
plt.show()
```

```python
# 4. Monthly Charges Density vs Churn
plt.figure(figsize=(8,5))
sns.kdeplot(data=df, x='MonthlyCharges', hue='Churn', fill=True, common_norm=False, palette='crest')
plt.title('Are higher Monthly Charges driving Churn?')
plt.show()
```
